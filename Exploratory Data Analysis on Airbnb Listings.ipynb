```python
total_cells = np.product(df.shape)
total_missing = df.isnull().sum()

# percent of data that is missing
(total_missing/total_cells) * 100
```

```python
numerical_columns = df.select_dtypes(include=np.number).columns
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].median())

# Step 2: Replace missing values in categorical columns with mode
categorical_columns = df.select_dtypes(include='object').columns
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])
```

```python
room_type_counts = df ['room_type'].value_counts()
print("Count of Various Room Types Available:")
print(room_type_counts)
```

The entire place is the most common type of accommodation available. It indicates that 182,005 listings are for entire homes or apartments, where guests have the entire space to themselves without sharing it with others.

```python
avg_price_neighborhood = df.groupby('neighbourhood')['price'].mean().sort_values(ascending=False)
print("Average price per neighborhood:")
print(avg_price_neighborhood)
```

Sao Cristovao has the highest average price for accommodations at approximately 8,120.69 euros. This suggests it is a very upscale area, possibly with luxury listings.

```python
df["host_since"]=pd.to_datetime(df["host_since"])
df.info()
```

```python
paris_listings=(df.query("city== 'Paris'").loc[:,["host_since","neighbourhood","city","accommodates","price"]])
paris_listings.info()
```

```python
paris_listings_neighbourhood= (paris_listings.groupby("neighbourhood").agg({"price":"mean"}).sort_values("price"))
paris_listings_neighbourhood.tail(10)
```

Elysee is the most expensive neighborhood in Paris, with an average listing price of 210.54 euros. 

```python
paris_listings_neighbourhood.head(10)
```

Menilmontant is the most affordable neighborhood in Paris, with an average listing price of 74.94 euros. This area is likely more budget-friendly and may appeal to travelers seeking economical accommodations.

```python
paris_listings_accommodate=(paris_listings.query("neighbourhood == 'Elysee'").groupby("accommodates").agg({"price":"mean"}).sort_values("price"))
paris_listings_accommodate.head()
```

This highlights how the cost of accommodations in Elysee varies significantly based on the number of people they can host. As the number of accommodates increases, the average price also increases.
